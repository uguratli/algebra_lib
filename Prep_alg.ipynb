```python
indice_s1=["a","b","c","d","e","f"]
indice_s2=["c d","e f","a b","b c","a d"]
indice_q=["\\alpha","\\beta"]
opt_s0=[0]
opt_s1=[0]
opt_s2=[0]
opt_q1=[1]
commutations_=[]
cadabra_alg=open("test_algebra.txt","w")
cadabra_alg.write("{")
for commutation in commutations:
    commutation=commutation.replace("M_","J_").replace("K_","L_").replace("D_{+}","D").replace("D_{-}","E") 
    Lhs,Rhs=commutation.split("=")
    Rhs=Rhs.replace("\n","").replace("\r","")
    Lhs=Lhs.replace("\left[","[").replace("\\right]","]")
    p,q=re.search("\[(.*)\]",Lhs).group(1).split(",")
    q = q[:0] + q[1:]
    if len(p)>1:
        indice=re.search("_(.*)",p).group(1).replace("{","").replace("}","")
        if indice in indice_s1 and p[0] not in opt_s1:
            opt_s1.append(p[0])
        elif indice in indice_s2 and p[0] not in opt_s2:
            opt_s2.append(p[0])
    else:
        if p[0] not in opt_s0:
            opt_s0.append(p[0])
    if len(q)>1:
        indice=re.search("_(.*)",q).group(1).replace("{","").replace("}","")
        if indice in indice_s1 and q[0] not in opt_s1:
            opt_s1.append(q[0])
        elif indice in indice_s2 and q[0] not in opt_s2:
            opt_s2.append(q[0])
    else:
        if q[0] not in opt_s0:
            opt_s0.append(q[0])
    commutations_.append((p[0],q[0]))
    cadabra_alg.write("\commutator{%s}{%s}->%s""\commutator{%s}{%s}->-\commutator{%s}{%s}," %(p,q,Rhs,q,p,p,q))
print(opt_s2)
print(opt_s1)
print(opt_s0)
print(commutations_)
opt_set=opt_s0[1:]+ opt_s1[1:]+opt_s2[1:]
for p,q in combinations_with_replacement(opt_set,2):
    i,j=0,0
    if (p,q) in commutations_ or (q,p) in commutations_:
        continue
    else:
        if p in opt_s2:
            p=p+"_{"+indice_s2[i]+"}"
            i+=1
        elif p in opt_s1:
            p=p+"_"+indice_s1[j]
            j+=1
        if q in opt_s2:
            q=q+"_{"+indice_s2[i]+"}"
            i+=1
        elif q in opt_s1:
            q=q+"_"+indice_s1[j]
            j+=1
        cadabra_alg.write("\commutator{%s}{%s}-> 0,"
                        "\commutator{%s}{%s}->-\commutator{%s}{%s}," %(p,q,q,p,p,q))
cadabra_alg.write("}")
```
